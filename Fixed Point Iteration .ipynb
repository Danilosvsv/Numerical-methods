## Fixed Point Iteration 

A fixed point of a function is a number whose value of a function is not changed by applying the function.

```python
from numpy import *
import numpy as np 

def fixed_point(p0,TOL,N0):
    i = 1
    g = lambda x: np.sqrt(10/(4+x))
    while(i <= N0):
        p = g(p0)
        if(np.absolute(p-p0) < TOL):
            print(p)
            break
        i = i+1
        p0 = p
```

```python
fixed_point(1.5,0.00001,300)
```

One way to visualize the result obtained after the application of the method is to visualize the function graph, because in this way it is possible to observe that the fixed point approximation satisfies the condition p = g(p).

```python
import matplotlib.pyplot as plt  

# The result presented in the approximate solution, can be visualized through the use of the graphical view.

# Two meshes are defined: the first is defined arbitrarily while the second is defined close to the approximate solution.

x1 = np.linspace(0,3)
x2 = np.linspace(1.3,1.4)

plt.figure()
plt.subplot()
plt.plot(x1, g(x1))  
plt.title('Graph of g(x) =  np.sqrt(10/(4+x))')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()

plt.figure()
plt.subplot()
plt.plot(x2, g(x2))  
plt.title('Graph of g(x) = np.sqrt(10/(4+x)) in an interval close to the root.')
plt.xlabel('x')
plt.ylabel('g(x)')

plt.grid()
plt.show()
```

On the graphs, at first the graph of g (x) is plotted and then we reduced the search interval on the second graph to observe graphically if the approximate point obtained is a fixed point. Once plotted, it is possible to infer whether the condition $p = g(p)$ is satisfied. In the case of figure 2, note that if $p = 1.3652...$, then $g(p) \approx 1.365...$. It is worth mentioning that it is quite possible that there is a numerical or machine error associated, however, this is another topic to be explored.
